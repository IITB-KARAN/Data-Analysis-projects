```python
import numpy as np
import matplotlib.pyplot as plt

def normal():
    N = int(input("Enter the number of random samples (N): "))
    mu = float(input("Enter the mean: "))
    sigma = float(input("Enter the standard deviation: "))
    
    data = np.random.normal(loc=mu, scale=sigma, size=N)
    
    # Plot histogram
    plt.hist(data, bins=30,density=True,alpha=0.6, color='skyblue', edgecolor='black')
    
    from scipy.stats import norm
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, loc= mu, scale=sigma)
    plt.plot(x, p, 'r',linewidth=2)
    
    plt.title('Histogram plot')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

#as i am increasing no. of samples the histogram plot begins to resemble the normal distribution more and more
normal()
normal()
normal()
```

```python
#part2 Q3
import pandas as pd
import numpy as np
import scipy.stats as sc
import matplotlib.pyplot as plt

Auto = pd.read_csv('Auto.csv',na_values =['?'])
Auto=Auto.dropna()


displacement = Auto["displacement"]
horsepower = Auto["horsepower"]


def manual(x,n):
    mean_val = sum(x) / n
    
    # Median
    sorted_x = np.sort(x)
    if n % 2 == 0:
        median_val = (sorted_x[n//2 - 1] + sorted_x[n//2]) / 2
    else:
        median_val = sorted_x[n//2]
    
    # Standard deviation (sample std, denominator n-1)
    variance = sum((xi - mean_val) ** 2 for xi in x) / (n - 1)
    std_val = variance ** 0.5
    
    # Skewness 
    skewness_val = (sum((xi - mean_val) ** 3 for xi in x) / n) / (std_val ** 3)
    
    kurtosis_val = (sum((xi - mean_val) ** 4 for xi in x) / n) / (std_val ** 4) - 3
    
    print("Manual Statistics value")
    print("Mean:",mean_val)
    print("Median:",median_val)
    print("Std Dev:", std_val)
    print("Skewness:", skewness_val)
    print("Kurtosis:", kurtosis_val)
    print()




manual(displacement,len(displacement))

print("Displacement ")
print("Mean:", np.mean(displacement))
print("Median:", np.median(displacement))
print("Std Dev:", np.std(displacement, ddof=1))
print("Skewness:", sc.skew(displacement))
print("Kurtosis:", sc.kurtosis(displacement))
print()

manual(horsepower,len(horsepower))
print("Horsepower ")
print("Mean:", np.mean(horsepower))
print("Median:", np.median(horsepower))
print("Std Dev:", np.std(horsepower, ddof=1))
print("Skewness:", sc.skew(horsepower))
print("Kurtosis:", sc.kurtosis(horsepower))
print()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(displacement, bins=30, color="skyblue", edgecolor="black")
plt.title("Histogram of Displacement")
plt.xlabel("Displacement")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(horsepower, bins=30, color="lightgreen", edgecolor="black")
plt.title("Histogram of Horsepower")
plt.xlabel("Horsepower")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

population_size = 100000
population = np.random.uniform(0, 100, population_size)

a, b = 0, 100
mu = (a + b) / 2
sigma = ((b - a)**2 / 12) ** 0.5

print("Theoretical mean:", mu)
print("Theoretical std:", sigma)


n = 30  
N = 10000

sample_means = np.zeros(N)

for i in range(N):
    sample = np.random.choice(population, size=n, replace=False)
    sample_means[i]=(np.mean(sample))



plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(population, bins=50, color="skyblue", edgecolor="black")
plt.title("Histogram of Population (Uniform [0,100])")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Histogram of sample means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=50, color="lightgreen", edgecolor="black")
plt.title("Histogram of Sample Means (n=30, 10,000 trials)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
print("Sample means - Mean:", np.mean(sample_means))
print("Sample means - Std:", np.std(sample_means, ddof=1))
```
